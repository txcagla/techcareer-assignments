```python
# gerekli kütüphaneleri import etme
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
# veri setini okuma
data = pd.read_csv('/content/50_Startups.csv')
```

```python
# ilk beş satırı görüntüleme
data.head()
```

```python
# 1. GÖREV: R&D Harcaması ve Kâr Arasındaki İlişki (Scatter Plot)

plt.figure(figsize=(8,6))
plt.scatter(data['R&D Spend'], data['Profit'], color='green')
plt.title('R&D Spending vs Profit', fontsize=14)
plt.xlabel('R&D Spend ($)', fontsize=12)
plt.ylabel('Profit ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 2. GÖREV: Yönetim Harcamaları ve Kâr Arasındaki İlişki (Scatter Plot)

plt.figure(figsize=(8,6))
plt.scatter(data['Administration'], data['Profit'], color='purple')
plt.title('Administration Spending vs Profit', fontsize=14)
plt.xlabel('Administration Spend ($)', fontsize=12)
plt.ylabel('Profit ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 3. GÖREV: Eyaletlere Göre Ortalama Kâr (Bar Chart)

# Her eyaletin ortalama kârını hesaplama
state_profit = data.groupby('State')['Profit'].mean()

plt.figure(figsize=(8,6))
state_profit.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Average Profit by State', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Average Profit ($)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# 4. GÖREV: Harcama Türlerinin Karşılaştırması (Boxplot)

plt.figure(figsize=(8,6))
plt.boxplot([data['R&D Spend'], data['Administration'], data['Marketing Spend']],
            labels=['R&D Spend', 'Administration', 'Marketing Spend'],
            patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red', linewidth=2))
plt.title('Comparison of Different Spending Types', fontsize=14)
plt.ylabel('Spending Amount ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
