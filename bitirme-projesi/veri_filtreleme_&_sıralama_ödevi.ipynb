```python
# gerekli kütüphaneleri import etme
import pandas as pd
import numpy as np
```

```python
# veri setini yükleme
data = pd.read_csv('/content/country.csv')
```

```python
#ilk beş satırı görüntüleme
data.head()
```

```python
# 1. GÖREV: Nüfusa göre azalan sırada ülkeleri listeleme

# 'Population' sütununa göre azalan (descending) sırala
population_desc = data.sort_values(by='Population', ascending=False)
print("Nüfusa göre azalan sıralama:")
display(population_desc[['Country', 'Population']].head(10))
```

```python
# 2. GÖREV: GDP per capita’ya göre artan sırada ülkeleri listeleme

gdp_asc = data.sort_values(by='GDP ($ per capita)', ascending=True)
print("\nGDP per capita’ya göre artan sıralama:")
display(gdp_asc[['Country', 'GDP ($ per capita)']].head(10))
```

```python
# 3. GÖREV: Nüfusu 10 milyonun üzerinde olan ülkeleri seçme

over_10m = data[data['Population'] > 10_000_000]
print("\nNüfusu 10 milyonun üzerinde olan ülkeler:")
display(over_10m[['Country', 'Population']])
```

```python
# 4. GÖREV: Literacy oranına göre en yüksek ilk 5 ülkeyi seçme

# Önce Literacy sütunundaki eksik değerleri düşür (NaN varsa)
top5_literacy = data.dropna(subset=['Literacy (%)']).sort_values(by='Literacy (%)', ascending=False).head(5)

print("\nEn yüksek okuryazarlık oranına sahip ilk 5 ülke:")
display(top5_literacy[['Country', 'Literacy (%)']])
```

```python
# 5. GÖREV: GDP per capita > 10.000 olan ülkeleri filtreleme

rich_countries = data[data['GDP ($ per capita)'] > 10_000]
print("\nGDP per capita 10.000’den büyük ülkeler:")
display(rich_countries[['Country', 'GDP ($ per capita)']])
```

```python
# 6. GÖREV: Nüfus yoğunluğu en yüksek ilk 10 ülkeyi bulma

dense_countries = data.sort_values(by='Pop. Density (per sq. mi.)', ascending=False).head(10)
print("\nEn yüksek nüfus yoğunluğuna sahip ilk 10 ülke:")
display(dense_countries[['Country', 'Pop. Density (per sq. mi.)']])
```
