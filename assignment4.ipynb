Numpy Bölümü: Soru 1

```python
import numpy as np

# 5x5 boyutunda rastgele (0-100 arasında) tam sayılardan matris oluşturma
matris = np.random.randint(0, 101, size=(5,5))
print("Oluşturulan Rastgele Matris:\n", matris)

# ortalama, standart sapma ve varyans hesaplama
ortalama = np.mean(matris)
standart_sapma = np.std(matris)
varyans = np.var(matris)

print("\nOrtalama:", ortalama)
print("Standart Sapma:", standart_sapma)
print("Varyans:", varyans)

# en büyük ve en küçük değeri bulma
max_deger = np.max(matris)
min_deger = np.min(matris)

print("\nEn büyük değer:", max_deger)
print("En küçük değer:", min_deger)

# köşegen elemanlarının toplamını bulma
kosegen_toplam = np.trace(matris)
print("\nKöşegen elemanlarının toplamı:", kosegen_toplam)
```

Numpy Bölümü: Soru 2

```python
import numpy as np

# 1000 öğrencinin sınav puanlarını simüle etme
puanlar = np.random.normal(loc=50, scale=17, size=1000)
puanlar = np.clip(puanlar, 0, 100)

# ortalama, medyan ve standart sapma hesaplama
ortalama = np.mean(puanlar)
medyan = np.median(puanlar)
standart_sapma = np.std(puanlar)

print("Ortalama:", ortalama)
print("Medyan:", medyan)
print("Standart Sapma:", standart_sapma)

# 50’den düşük alan öğrenci sayısını bulma
dusuk_sayisi = np.sum(puanlar < 50)
print("\n50'den düşük alan öğrenci sayısı:", dusuk_sayisi)
```

Pandas Bölümü

```python
import pandas as pd

# 1. Veri setini DataFrame olarak oluşturma
data = {
    "Öğrenci": ["Ali", "Ayşe", "Mehmet", "Zeynep", "Ahmet"],
    "Yaş": [20, 21, 19, 22, 20],
    "Bölüm": ["Bilgisayar", "Fizik", "Kimya", "Bilgisayar", "Fizik"],
    "Matematik": [70, 60, 80, 90, 55],
    "Fizik": [65, 75, 70, 85, 60],
    "Kimya": [80, 85, 65, 95, 70]
}

df = pd.DataFrame(data)
print("Orijinal DataFrame:\n", df)

# 2. Her ders için ortalama puan
print("\nHer dersin ortalaması:\n", df[["Matematik","Fizik","Kimya"]].mean())

# 3. En yüksek matematik notunu alan öğrenci
max_math = df.loc[df["Matematik"].idxmax()]
print("\nEn yüksek Matematik notunu alan öğrenci:\n", max_math)

# 4. Her öğrencinin not ortalamasını hesaplayan yeni sütun
df["Ortalama"] = df[["Matematik","Fizik","Kimya"]].mean(axis=1)
print("\nOrtalama sütunu eklenmiş DataFrame:\n", df)

# 5. Bölümlere göre gruplayarak ortalama başarıları hesaplama
print("\nBölümlere göre ortalama başarılar:\n", df.groupby("Bölüm")[["Matematik","Fizik","Kimya","Ortalama"]].mean())

# 6. Ortalaması 70’in üzerinde olan öğrenciler
print("\nOrtalaması 70'in üzerinde olan öğrenciler:\n", df[df["Ortalama"] > 70])
```

Ek Görev

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.hist(df["Matematik"], bins=5, color="skyblue", edgecolor="black")
plt.title("Matematik Dağılımı")
plt.xlabel("Not")
plt.ylabel("Öğrenci Sayısı")

plt.subplot(1, 3, 2)
plt.hist(df["Fizik"], bins=5, color="lightgreen", edgecolor="black")
plt.title("Fizik Dağılımı")
plt.xlabel("Not")
plt.ylabel("Öğrenci Sayısı")

plt.subplot(1, 3, 3)
plt.hist(df["Kimya"], bins=5, color="salmon", edgecolor="black")
plt.title("Kimya Dağılımı")
plt.xlabel("Not")
plt.ylabel("Öğrenci Sayısı")

plt.tight_layout()
plt.show()

bolum_ort = df.groupby("Bölüm")[["Matematik","Fizik","Kimya","Ortalama"]].mean()

bolum_ort["Ortalama"].plot(kind="bar", color="orange", edgecolor="black")
plt.title("Bölümlere Göre Ortalama Başarılar")
plt.ylabel("Ortalama Not")
plt.xlabel("Bölüm")
plt.show()
```
